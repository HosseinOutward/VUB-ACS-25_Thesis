```
import numpy as np
import matplotlib.pyplot as plt
```

```
side_info_data = np.random.normal(0, 1, 1_000_000, ).astype(np.float32)
y = side_info_data + np.random.normal(0, np.sqrt(0.1), 1_000_000, ).astype(np.float32)
side_info_data = [side_info_data]
temp = np.percentile(y, [0.0003, 99.9997])
temp = ((y >= temp[0]) * (y <= temp[1]))
y = y[temp]
side_info_data = [a[temp] for a in side_info_data]

y/=4.5
1/np.mean(np.abs(y)**2)
```

```
noise_power = 0.1
plane_setting = '44'
```

```
wyner_ziv_bound_f = lambda d: 1/2 * np.log2(noise_power/d) if 0<d<noise_power else None
latic_error = 1.53
```

```
ld_list, bit_rate_list, mae_db_list, mape_db_list = np.load('lambda_distortion_data_44_0,1.npz')['arr_0']
```

```
fig, ax = plt.subplots(1,2, figsize=(22, 10))
for i, metric in enumerate([mae_db_list, mape_db_list]):

    D=np.linspace(noise_power*1e-4, noise_power, 1000)
    if i==1:
        D = D/np.mean(np.abs(y))
    bit_rate_wz_bound=[wyner_ziv_bound_f(d) for d in D]
    D = 10 * np.log10(D)
    ax[i].plot(bit_rate_wz_bound, D, linewidth=1)
    ax[i].plot(bit_rate_wz_bound, D+latic_error, linewidth=1, linestyle=':')

    ax[i].plot(bit_rate_list, metric, linewidth=1, linestyle='--')
    ax[i].scatter(bit_rate_list, metric, c=ld_list,)
    ax[i].set_xlabel('Bit Rate (bpp)')
    ax[i].set_ylabel('Distortion (dB)')
    ax[i].set_title(f'Plane {plane_setting} - {["MAE", "MAPE"][i]}')
    ax[i].grid(True)

```
